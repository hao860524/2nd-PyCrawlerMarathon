```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
import re

rows = table.find_all('td')

data = {}

index = []
r_str = ""

for r in rows:
    if re.search(re.compile("^[A-Z]+\w+$"), r.text):
        r_str = str(r.text)
    elif re.search(re.compile("^\d+\.*\d*$"), r.text):
        if not r_str in data:
            data[r_str] = []
            data[r_str].append(r.text)
        else:
            data[r_str].append(r.text)
    elif re.search(re.compile("^\d+\/\d+$"), r.text):
        if not r.text in index:
            index.append(r.text)

print(index)
print(data)
```

```python
import pandas as pd
d = pd.DataFrame(data, index=index)
pd.DataFrame(d)
```
