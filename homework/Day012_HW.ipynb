### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
news = soup.find(class_="part_list_2").find_all('h3')
for new in news:
    print(new.find(class_="date").text,new.a.text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
classifications = {}
classes = []

for new in news:
    tag = new.em['class'][1][2:]
    if not (tag in classifications.keys()):
        classes.append(str((tag)))
        classifications[tag] = []
    classifications[tag].append({'Time' : new.find(class_="date").text , 'Title' :new.a.text.replace(u'\u3000',u' ')})

for c in classes:
    print(c)
    for each in classifications[c]:
        print(each)
```

### ③ 哪一個類別的文章最多

```python
name,num = [],0

for c in classes:
    if num < len(classifications[c]):
        num = len(classifications[c])
        
for c in classes:
    if len(classifications[c]) == num:
        name.append(c)

for n in name: print(n,end=' ')
    
print("{}篇".format(num))
```
