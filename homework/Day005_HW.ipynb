```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

import json

data = json.loads(r.text)
# print(type(json.loads(r.text)))
# print(type(data[0]))

print("{}ç­†".format(len(data)))
print("{}å€‹æ¬„ä½\n{}".format(len(data[0]),data[0].keys()))
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

for d in data:
    print("æ¨™é¡Œï¼š{title}\nè²¼æ–‡æ™‚é–“ï¼š{time}\nç•™è¨€äººæ•¸ï¼š{commentCount} ğŸ’¬  æŒ‰è®šäººæ•¸ï¼š{likeCount} ğŸ‘\n\n".
          format(title = d['title'], time = d['createdAt'], commentCount = d['commentCount'], likeCount = d['likeCount']))
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

class Article:
    def __init__(self, category, pupular):
        self.pupular = pupular
        self.category = category
        self.response = json.loads(requests.get('https://www.dcard.tw/_api/forums/'+category+'/posts?popular='+pupular).text)
    
    def re_response(self):
        return self.response
    
    def avg_commentCount(self):
        cc = 0
        for article in self.response:
            cc += article['commentCount']
        return cc/len(self.response)
        
    def avg_likeCount(self):
        lc = 0
        for article in self.response:
            lc += article['likeCount']
        return lc/len(self.response)


popular = Article("pet","true")

unpopular = Article("pet","false")


print("ç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€ç‚º:{}å€‹ ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€ç‚º:{}äºº".format(popular.avg_commentCount(),popular.avg_likeCount()))
print("éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€ç‚º:{}å€‹ ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€ç‚º:{}äºº".format(unpopular.avg_commentCount(),unpopular.avg_likeCount()))

```
